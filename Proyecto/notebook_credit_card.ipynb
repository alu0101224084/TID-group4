```
#fraud vs. normal transactions 
def compareFraudTransactions(dataset):
    counts = dataset.Class.value_counts()
    normal = counts[0]
    fraudulent = counts[1]
    perc_normal = (normal / (normal + fraudulent)) * 100
    perc_fraudulent = (fraudulent / (normal + fraudulent)) * 100
    print('Se realizaron {} transacciones no fraudulentas ({:.2f}%) y {} transacciones fraudulentas ({:.2f}%).'.format(normal, perc_normal, fraudulent, perc_fraudulent))

compareFraudTransactions(cc_df)

# Remove duplicates
cc_df.drop_duplicates(keep='first', inplace=True)
print('Despues de quitar duplicados:')
compareFraudTransactions(cc_df)

# There is no difference at all
```

```
from sklearn.preprocessing import RobustScaler

def scaleDataset(dataset):
    rob_scaler = RobustScaler()
    scaled_amount = rob_scaler.fit_transform(dataset['Amount'].values.reshape(-1,1))
    scaled_time = rob_scaler.fit_transform(dataset['Time'].values.reshape(-1,1))

    scaled_dataset = dataset.copy()
    scaled_dataset.drop(['Time','Amount'], axis=1, inplace=True)

    scaled_dataset.insert(0, 'scaled_amount', scaled_amount)
    scaled_dataset.insert(1, 'scaled_time', scaled_time)

    return scaled_dataset

scaled_dataset = scaleDataset(cc_df)
# TODO quitar esta pirateria
cc_df = scaled_dataset
```

```
from sklearn.metrics import recall_score, precision_score, accuracy_score, f1_score, roc_auc_score
from sklearn.metrics import confusion_matrix, mean_absolute_error, mean_squared_error
import seaborn as sns

def printScores(y_test, y_pred, y_pred_prob):
    print("recall score: ", recall_score(y_test, y_pred))
    print("precision score: ", precision_score(y_test, y_pred))
    print("f1 score: ", f1_score(y_test, y_pred))
    print("accuracy score: ", accuracy_score(y_test, y_pred))
    print("ROC AUC: {}".format(roc_auc_score(y_test, y_pred_prob[:,1])))
    print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
    print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
    print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred)))
    class_names = ['not_fraud', 'fraud']
    matrix = confusion_matrix(y_test, y_pred)
    # Create pandas dataframe
    dataframe = pd.DataFrame(matrix, index=class_names, columns=class_names)
    # Create heatmap
    sns.heatmap(dataframe, annot=True, cbar=None, cmap="Blues", fmt = 'g')
    plt.title("Confusion Matrix"), plt.tight_layout()
    plt.ylabel("True Class"), plt.xlabel("Predicted Class")
    plt.show()

def getPredictions(clf, X_train, y_train, X_test):
    # fit it to training data
    clf.fit(X_train,y_train)
    # predict using test data
    y_pred = clf.predict(X_test)
    # Compute predicted probabilities: y_pred_prob
    y_pred_prob = clf.predict_proba(X_test)
    return y_pred, y_pred_prob
```

```
# DOWNSAMPLING

import imblearn
from imblearn.under_sampling import RandomUnderSampler
from sklearn.linear_model import LogisticRegression

def downsample(dataset):
    undersample = RandomUnderSampler(sampling_strategy=0.8, random_state=1)

    cols = dataset.columns.tolist()
    cols = [c for c in cols if c not in ["Class"]]
    target = "Class"

    #define X and Y
    X = dataset[cols]
    y = dataset[target]

    #undersample
    X_under, y_under = undersample.fit_resample(X, y)
    return X_under, y_under

def showClassDifferences(original_df, final_df):
    fig, axs = plt.subplots(ncols=2, figsize=(13,4.5))
    sn.countplot(x="Class", data=original_df, ax=axs[0])
    sn.countplot(x="Class", data=final_df, ax=axs[1])
    fig.suptitle("Class repartition before and after")
    a1=fig.axes[0]
    a1.set_title("Before")
    a2=fig.axes[1]
    a2.set_title("After")

X_under, y_under = downsample(train_dataset)
downsampled_train_y = pd.DataFrame(y_under, columns = ['Class'])
downsampled_train_dataset = pd.concat([X_under, y_under], axis=1)

showClassDifferences(cc_df, downsampled_train_y)
compareFraudTransactions(downsampled_train_y)
```

```
# REMOVING OUTLIERS

def removeOutliers(dataset):
    no_outliers = pd.DataFrame(dataset.copy())
    cc_corr = dataset.corr()

    # Removing Outliers from high-correlation features

    cols = cc_corr.Class.index[:-1]

    # For each feature correlated with Class...
    for col in cols:
        # If absolute correlation value is more than X percent...
        correlation = cc_corr.loc['Class',col]
        if np.absolute(correlation) > 0.1:
            
            # Separate the classes of the high-correlation column
            nonfrauds = no_outliers.loc[no_outliers.Class==0,col]
            frauds = no_outliers.loc[no_outliers.Class==1,col]

            # Identify the 25th and 75th quartiles
            all_values = no_outliers.loc[:,col]
            q25, q75 = np.percentile(all_values, 25), np.percentile(all_values, 75)
            # Get the inter quartile range
            iqr = q75 - q25
            # Smaller cutoffs will remove more outliers
            cutoff = iqr * 7
            # Set the bounds of the desired portion to keep
            lower, upper = q25 - cutoff, q75 + cutoff
            
            # If positively correlated...
            # Drop nonfrauds above upper bound, and frauds below lower bound
            if correlation > 0: 
                no_outliers.drop(index=nonfrauds[nonfrauds>upper].index,inplace=True)
                no_outliers.drop(index=frauds[frauds<lower].index,inplace=True)
            
            # If negatively correlated...
            # Drop nonfrauds below lower bound, and frauds above upper bound
            elif correlation < 0: 
                no_outliers.drop(index=nonfrauds[nonfrauds<lower].index,inplace=True)
                no_outliers.drop(index=frauds[frauds>upper].index,inplace=True)
            
    print('\nData shape before removing outliers:', dataset.shape)
    print('\nCounts of frauds VS non-frauds in previous data:')
    print(dataset.Class.value_counts())
    print('-'*40)
    print('-'*40)
    print('\nData shape after removing outliers:', no_outliers.shape)
    print('\nCounts of frauds VS non-frauds in new data:')
    print(no_outliers.Class.value_counts())
    return dataset

no_outliers_cc = removeOutliers(downsampled_train_dataset)
```

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import log_loss, accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from datetime import datetime
```

## PerceptrÃ³n Simple
Entrenamos el modelo PerceptrÃ³n Simple con 30 iteraciones, que nos informarÃ¡ del punto hasta el cual el problema es separable linealmente.

```
def simplePerceptron(input_train, input_valid, target_train, target_valid):
    max_iter = 30
    per = Perceptron(max_iter=max_iter, shuffle=False, random_state=1, verbose=True)
    per.fit(input_train, target_train)
    predict_train = per.predict(input_train)
    predict_valid = per.predict(input_valid)
    print("Train accuracy: %.3f%%" % (accuracy_score(target_train, predict_train) * 100))
    print("Valid accuracy: %.3f%%" % (accuracy_score(target_valid, predict_valid) * 100))

perceptron_X_train, perceptron_X_valid, perceptron_y_train, perceptron_y_valid = train_test_split(X_dropped, y_dropped, test_size=0.3, random_state=1)
simplePerceptron(perceptron_X_train, perceptron_X_valid, perceptron_y_train, perceptron_y_valid)
```

## PerceptrÃ³n Multicapa

```
def MLP_train_valid(mlp, input_train, target_train, input_valid, target_valid, max_iter, valid_cycles, verbose):
    """
    Train and valid MLP every valid_cycles iterations
    """
    classes = np.unique(target_train)
    loss_valid = []
    for i in range(int(max_iter/valid_cycles)):
        for j in range(valid_cycles):
            out = mlp.partial_fit(input_train, target_train, classes)
            # Calculate loss function of valid set
        last_lost_valid = log_loss(target_valid, mlp.predict_proba(input_valid))
        loss_valid.append(last_lost_valid)
        if verbose:
            print("Iteration %d, train loss = %.8f, valid loss = %.8f" %
                  (mlp.n_iter_, mlp.loss_, last_lost_valid))
        if early_stopping and (i > 0) and (last_lost_valid > loss_valid[-2]): # Early stopping
            if verbose:
                print("Early stopping: Validation score did not improve")
            break
    if verbose: 
        print(out)
        print("Visualizing the Cost Function Trajectory")
        plt.plot(range(1, len(mlp.loss_curve_)+1), mlp.loss_curve_, label='Train loss')
        plt.plot(range(valid_cycles,len(loss_valid)*valid_cycles+valid_cycles,valid_cycles), loss_valid, '-o', label='Valid loss')
        plt.xlabel('number of iterations')
        plt.ylabel('loss function')
        plt.legend(loc='upper right')
        plt.show()

n_hidden = 60
max_iter = 300
learning_rate_init = 0.001
valid_cycles = 5
early_stopping = True

print("Learning a MLP with %d hidden neurons, %d maximum number of iterations and %.8f learning rate ..." % (n_hidden, max_iter, learning_rate_init))

mlp = MLPClassifier(hidden_layer_sizes=(n_hidden,), learning_rate_init=learning_rate_init, shuffle=False, random_state=0, verbose=False)

MLP_train_valid(mlp, perceptron_X_train, perceptron_y_train, perceptron_X_valid, perceptron_y_valid, max_iter, valid_cycles, True)
```

```
predict_train = mlp.predict(perceptron_X_train)
predict_valid = mlp.predict(perceptron_X_valid)
print("Train accuracy: %.3f%%" % (accuracy_score(perceptron_y_train, predict_train) * 100))
print("Valid accuracy: %.3f%%" % (accuracy_score(perceptron_y_valid, predict_valid) * 100))
print("Train confusion matrix:")
print(confusion_matrix(perceptron_y_train, predict_train))
print("Valid confusion matrix:")
print(confusion_matrix(perceptron_y_valid, predict_valid))
print("Train classification report:")
print(classification_report(perceptron_y_train, predict_train))
print("Valid classification report:")
print(classification_report(perceptron_y_valid, predict_valid))
```

OptimizaciÃ³n ratio de aprendizaje

```
tests_learning_rate_init = [0.001, 0.0015, 0.002, 0.0025, 0.003, 0.005, 0.01, 0.05, 0.1]
activation = 'relu'
random_state = 1
now = datetime.now()
loss_curves =  []
for lr in tests_learning_rate_init:
   mlp = MLPClassifier(hidden_layer_sizes=(n_hidden,), learning_rate_init=lr, shuffle=False, random_state=random_state, verbose=False, activation=activation)
   MLP_train_valid(mlp, perceptron_X_train, perceptron_y_train, perceptron_X_valid, perceptron_y_valid, max_iter, valid_cycles, False)
   loss_curves.append(mlp.loss_curve_)
for (lr, loss_curve) in zip(tests_learning_rate_init, loss_curves):
   plt.plot(range(1, len(loss_curve)+1), loss_curve, label='learning rate = ' + str(lr))
plt.xlabel('number of iterations')
plt.ylabel('loss function')
plt.legend(loc='upper right')
plt.show()
```

## OptimizaciÃ³n arquitectura

```
tests_n_hidden = [10, 30, 50, 70, 90, 110, 130, 150, 170, 190]
n_reps = 10
activation = 'relu'
# activation = 'logistic'
learning_rate_init = 0.0025
now = datetime.now()
best_mlp = []
best_acc = 0.0
accs_train = []
accs_valid = []
for n_hidden in tests_n_hidden:
   max_acc_train = max_acc_valid = 0.0
   for random_state in range(n_reps):
       mlp = MLPClassifier(hidden_layer_sizes=(n_hidden,), learning_rate_init=learning_rate_init, shuffle=False, random_state=random_state, verbose=False, activation=activation)
       MLP_train_valid(mlp, perceptron_X_train, perceptron_y_train, perceptron_X_valid, perceptron_y_valid, max_iter, valid_cycles, False)
       
       acc_train = accuracy_score(perceptron_y_train, mlp.predict(perceptron_X_train))
       acc_valid = accuracy_score(perceptron_y_valid, mlp.predict(perceptron_X_valid))
       print("Seed = %d, train acc = %.8f, valid acc = %.8f, iterations = %d" % (random_state, acc_train, acc_valid, len(mlp.loss_curve_)))
       if (max_acc_valid < acc_valid):
           max_acc_valid = acc_valid
           max_acc_train = acc_train
           if (acc_valid > best_acc):
               best_acc = acc_valid
               best_mlp = mlp
   accs_train.append(max_acc_train)
   accs_valid.append(max_acc_valid)
   print("Number hidden units = %i, train acc = %.8f, max valid acc = %.8f" % (n_hidden, max_acc_train, max_acc_valid))
print("Number of seconds for training: %d" % (datetime.now() - now).total_seconds())
print("Best MLP valid accuracy: %.8f%%" % (best_acc * 100))
print("Best MLP: ", best_mlp)
```

```
from neupy import algorithms, utils, init

# Modeling SOM
# ------------

SOM_X_train = perceptron_X_train
SOM_y_train = perceptron_y_train
SOM_X_valid = perceptron_X_valid
SOM_y_valid = perceptron_y_valid

max_iter_som = 55
grid_height = 20
grid_width = 20
distance = 'euclid'
learning_radius = 5
step = 0.5
reduce_step_after = max_iter_som - 5
std = 1.0
reduce_std_after = max_iter_som - 5
weight = init.Normal()

print("Learning %dx%d SOM with %d maximum number of iterations and ..." % (grid_height, grid_width, max_iter_som))

now = datetime.now()
# Random generator seed for NeuPy
utils.reproducible(0)

sofm = algorithms.SOFM(
    n_inputs = SOM_X_train.shape[1],
    features_grid = (grid_height, grid_width),
    distance = distance,
    weight = weight,
    learning_radius = learning_radius,
    reduce_radius_after = max_iter_som // learning_radius,  # 0 radius at end
    step = step,
    reduce_step_after = reduce_step_after,
    std = std,
    reduce_std_after = reduce_std_after,
    shuffle_data = False,
    verbose = True,
)

sofm.train(SOM_X_train, epochs=max_iter_som)
sofm_output_train = sofm.predict(SOM_X_train)
sofm_output_valid = sofm.predict(SOM_X_valid)
print("Number of seconds for training: %d" % (datetime.now() - now).total_seconds())
```

```
import matplotlib.gridspec as gridspec

def plot_prototypes_grid(grid_height, grid_width, weight):
    """
    Visualization prototypes of SOM grid
    """
    print("Building visualization of prototypes grid ...")
    grid = gridspec.GridSpec(grid_height, grid_width)
    grid.update(wspace=0, hspace=0)
    for row_id in range(grid_height):
        print("Progress: {:.2%}".format(row_id / grid_height))
        for col_id in range(grid_width):
            index = row_id * grid_width + col_id
            sample = weight[:,index]
            _ = plt.subplot(grid[index])
            _ = plt.imshow(sample.reshape((len(sample), 1)), cmap='PuBu')
            _ = plt.axis('off')
    plt.show()

plot_prototypes_grid(grid_height, grid_width, sofm.weight)
```

```
from itertools import product

def iter_neighbours(weights, hexagon=False):
    grid_height, grid_width = weights.shape

    hexagon_even_actions = ((-1, 0), (0, -1), (1, 0), (0, 1), (1, 1), (-1, 1))
    hexagon_odd_actions = ((-1, 0), (0, -1), (1, 0), (0, 1), (-1, -1), (1, -1))
    rectangle_actions = ((-1, 0), (0, -1), (1, 0), (0, 1))

    for neuron_x, neuron_y in product(range(grid_height), range(grid_width)):
        neighbours = []

        if hexagon and neuron_x % 2 == 1:
            actions = hexagon_even_actions
        elif hexagon:
            actions = hexagon_odd_actions
        else:
            actions = rectangle_actions

        for shift_x, shift_y in actions:
            neigbour_x = neuron_x + shift_x
            neigbour_y = neuron_y + shift_y

            if 0 <= neigbour_x < grid_height and 0 <= neigbour_y < grid_width:
                neighbours.append((neigbour_x, neigbour_y))

        yield (neuron_x, neuron_y), neighbours

def compute_heatmap(height, width, weight):
    heatmap = np.zeros((height, width))
    for (neuron_x, neuron_y), neighbours in iter_neighbours(weight):
        total_distance = 0

        for (neigbour_x, neigbour_y) in neighbours:
            neuron_vec = weight[neuron_x, neuron_y]
            neigbour_vec = weight[neigbour_x, neigbour_y]

            distance = np.linalg.norm(neuron_vec - neigbour_vec)
            total_distance += distance

        avg_distance = total_distance / len(neighbours)
        if neuron_y < height and neuron_x < width:
            heatmap[neuron_x, neuron_y] = avg_distance

    return heatmap

heatmap = compute_heatmap(grid_height, grid_width, sofm.weight)

plt.imshow(heatmap, cmap='magma', interpolation='nearest')

plt.axis('off')
plt.colorbar()
plt.show()
```

```
max_iter_per = 30

print("Learning a Perceptron with %d maximum number of iterations and ..." % max_iter_per)

per = Perceptron(max_iter=max_iter_per, shuffle=False, random_state=0, verbose=True)
per.fit(sofm_output_train, SOM_y_train)


# Intitial results
# ------------
print("Printing initial results")

predict_train = per.predict(sofm_output_train)
predict_valid = per.predict(sofm_output_valid)

print("Train accuracy: %.3f%%" % (accuracy_score(SOM_y_train, predict_train) * 100))
print("Valid accuracy: %.3f%%" % (accuracy_score(SOM_y_valid, predict_valid) * 100))

print("Train confusion matrix:")
print(confusion_matrix(SOM_y_train, predict_train))
print("Valid confusion matrix:")
print(confusion_matrix(SOM_y_valid, predict_valid))

print("Train classification report:")
print(classification_report(SOM_y_train, predict_train))
print("Valid classification report:")
print(classification_report(SOM_y_valid, predict_valid))
```

```
print("Architecture optimization")

# Test SOM with differents number of grid units and several repetitions
tests_grid_side = [2, 3, 4, 5, 6, 7, 10, 15, 20]
n_reps = 5

now = datetime.now()
best_sofm = []
best_per = []
best_acc = 0.0
accs_train = []
accs_valid = []
for grid_side in tests_grid_side:
    max_acc_train = max_acc_valid = 0.0
    for random_state in range(n_reps):
        utils.reproducible(random_state)
        sofm = algorithms.SOFM(n_inputs = SOM_X_train.shape[1], features_grid = (grid_side, grid_side), distance = distance, weight = weight, learning_radius = learning_radius, reduce_radius_after = max_iter_som // learning_radius, step = step, reduce_step_after = reduce_step_after, std = std, reduce_std_after = reduce_std_after, shuffle_data = False, verbose = False)
        sofm.train(SOM_X_train, epochs=max_iter_som)
        sofm_output_train = sofm.predict(SOM_X_train)
        sofm_output_valid = sofm.predict(SOM_X_valid)
        per = Perceptron(max_iter=max_iter_per, shuffle=False, random_state=random_state, verbose=False)
        _ = per.fit(sofm_output_train, SOM_y_train)
        acc_train = accuracy_score(SOM_y_train, per.predict(sofm_output_train))
        acc_valid = accuracy_score(SOM_y_valid,per.predict(sofm_output_valid))
        print("Seed = %d, train acc = %.8f, valid acc = %.8f" % (random_state, acc_train, acc_valid))
        if (max_acc_valid < acc_valid):
            max_acc_valid = acc_valid
            max_acc_train = acc_train
            if (acc_valid > best_acc):
                best_acc = acc_valid
                best_per = per
                best_sofm = sofm
    accs_train.append(max_acc_train)
    accs_valid.append(max_acc_valid)
    print("Grid size = %ix%i, train acc = %.8f, max valid acc = %.8f" % (grid_side, grid_side, max_acc_train, max_acc_valid))

print("Number of seconds for training: %d" % (datetime.now() - now).total_seconds())
print("Best CPN valid accuracy: %.8f%%" % (best_acc * 100))
print("Best SOM: ", best_sofm)
print("Best Perceptron: ", best_per)
```
